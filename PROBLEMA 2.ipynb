```python
df_wine = df_wine.rename(columns={'Unnamed: 0': 'index','country':'pais','description':'descripcion','price':'precio','province':'provincia'})
df_wine.head(1)
```

```python
dfconti = pd.read_csv('https://gist.githubusercontent.com/kintero/7d1db891401f56256c79/raw/a61f6d0dda82c3f04d2e6e76c3870552ef6cf0c6/paises.csv',sep=',')
dfconti = dfconti.rename(columns={'nombre':'pais'})
dfconti.head(2)
```

```python
#Columna continente
df_wine = pd.merge(df_wine, dfconti, how="inner", on="pais")
df_wine.drop([' name', ' nom',' iso2', ' iso3', ' phone_code'], axis=1, inplace=True)
df_wine.head(2)
```

```python
#Columna descuento
def descuentos(continente):
    if continente=='Europa':
        return 0.09
    elif continente=='Asia':
        return 0.10
    elif continente=='América':
        return 0.08
    elif continente=='África':
        return 0.05
    elif continente=='Oceanía':
        return 0.07
    else:
        return None

df_wine['descuento']=df_wine['continente'].apply(descuentos)
df_wine.head(2) 
```

```python
#Columna precio con descuento
df_wine['precio_final']=df_wine.precio*(1-df_wine.descuento)
df_wine.head(2) 
```

### Parte 3

Genere 4 reportes por agrupamiento de datos. Deberá elegir que reportes realizar
    - Ejemplo: Por contienente mostrar los vinos mejor puntuados
    - Ejemplo2: Promedio de precio de vino y cantidad de reviews obtenidos según pais. Ordenado de mejor a peor 

Al menos uno de estos datos agrupados deberán ser almacenados en excel o csv

```python
def agrupar(continente):
    return df_wine.loc[df_wine.groupby('continente')['points'].idxmax()]

vinos_continente = agrupar(df_wine['continente'])

vinos_continente.to_csv('MejoresPuntuados.csv', sep=';', encoding='utf-8', index=False)
vinos_continente
```

```python
points_precio = df_wine.groupby('pais').agg({'precio': 'mean', 'points': 'sum'})

points_precio =points_precio.sort_values(by='points', ascending=False)

points_precio.to_excel('Precio por pais.xlsx', index=False)

points_precio
```
